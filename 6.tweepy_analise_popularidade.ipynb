```python
#https://www.youtube.com/watch?v=_JfI96Qs9o8

#importa bibliotecas
import tweepy as tw
import pandas as pd
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
# Set up das credenciais
consumer_key = os.environ['api_key']
consumer_secret = os.environ['api_secret_key']
token = os.environ['beares_token']
access_token = os.environ['access_token']
access_token_secret = os.environ['access_token_secret']
```

```python
# Conecta com API do Twitter
client = tw.Client(token, consumer_key, consumer_secret, access_token, access_token_secret)
```

```python
# Busca tweets conforme query, por padr√£o busca 10 resultados, se quiser mais √© necess√°rio indicar (max 100)
resposta = client.search_recent_tweets(query="BBB", max_results=100)
```

```python
# Salva os tweets em vari√°vel pr√≥rpia
tweets = resposta.data
```

```python
# Checando como retornar apenas o texto do primeiro tweet
tweets[0].text
```

```python
# Imprime todos os textos dos tweets
for tweet in tweets:
    print(tweet.text)
```

```python
# Cria base de dados vazia onde ser√£o apensadas as linhas contendo informa√ß√µes sobre cita√ß√µes dos participantes no tweet
base = []
```

```python
# Salva cada tweet
for tweet in tweets:
    
    #para cada tweet, cria um vetor com 22 elementos (participantes) e atribui 0 para todos os valores + 1 para o texto + 1 para indicar se RT
    linha = [0 for j in range(24)]

    linha[0] = tweet.text

    texto = tweet.text
    # Se for RT, desconsiderar o que vem antes dos :
    if('RT' in texto):
        posicao = texto.find(':')
        texto = texto[posicao+2:]
        linha[23] = 1

    #Alterar coluna do participante para 1 quando ele for citado
    linha[1] = 1 if ('aline' in texto.lower() or 'wirley' in texto.lower()) else 0
    linha[2] = 1 if ('amanda' in texto.lower()) else 0
    linha[3] = 1 if ('cara de sapato' in texto.lower() or 'sapato' in texto.lower()) else 0
    linha[4] = 1 if ('bruna' in texto.lower() or 'griphao' in texto.lower()) else 0
    linha[5] = 1 if ('bruno' in texto.lower()) else 0
    linha[6] = 1 if ('cezar' in texto.lower() or 'c√©zar' in texto.lower()) else 0
    linha[7] = 1 if ('cristian' in texto.lower() or 'cris' in texto.lower()) else 0
    linha[8] = 1 if ('domitila' in texto.lower() or 'domi' in texto.lower()) else 0
    linha[9] = 1 if ('fred' in texto.lower() or 'boco roso' in texto.lower()) else 0
    linha[10] = 1 if ('fred nicacio' in texto.lower() or 'nic√°cio' in texto.lower() or 'conqueixo' in texto.lower()) else 0
    linha[11] = 1 if ('gabriel' in texto.lower()) else 0
    linha[12] = 1 if ('gabriel santana' in texto.lower() or 'mosca' in texto.lower()) else 0
    linha[13] = 1 if ('gustavo' in texto.lower() or 'gustagro' in texto.lower()) else 0
    linha[14] = 1 if ('key' in texto.lower()) else 0
    linha[15] = 1 if ('larissa' in texto.lower()) else 0
    linha[16] = 1 if ('marilia' in texto.lower() or 'mar√≠lia' in texto.lower()) else 0
    linha[17] = 1 if ('marvvila' in texto.lower()) else 0
    linha[18] = 1 if ('mc' in texto.lower() or 'guime' in texto.lower() or 'guim√™' in texto.lower()) else 0
    linha[19] = 1 if ('paula' in texto.lower()) else 0
    linha[20] = 1 if ('ricardo' in texto.lower()) else 0
    linha[21] = 1 if ('sarah' in texto.lower()) else 0
    linha[22] = 1 if ('tina' in texto.lower()) else 0


    base.append(linha)
```

```python
# Para visualizar melhor, transformar em dataframe
baseBBB = pd.DataFrame(base)

# nomear colunas
baseBBB.columns = ['texto', 'aline', 'amanda', 'sapato', 'bruna', 'bruno', 'cezar', 'cristian', 'domitila', 'fred', 'fred nicacio', 'gabriel', 'gabriel santana', 'gustavo', 'key', 'larissa', 'marilia', 'marvvila', 'guime', 'paula', 'ricardo', 'sarah', 'tina', 'RT']

display(baseBBB)
```

```python
#Elimina colunas desnecess√°ria
baseView = baseBBB.drop(['texto', 'RT'], axis=1)

#Para ver quem foi o mais citado
baseView.sum().sort_values(ascending=False)
```

```python
#Criar dataframe com as cita√ß√µes

citacoes = baseView.sum().sort_values(ascending=False).to_list()
participantes = baseView.sum().sort_values(ascending=False).index.to_list()

basePop = pd.DataFrame(zip(participantes, citacoes))
basePop.columns = ['participante', 'citacoes']
display(basePop)
```

```python
# Mostrar percentual
total_citacoes = basePop['citacoes'].sum()

basePop['%'] = basePop['citacoes'] / total_citacoes

basePop
```

```python
# Mostrar percentual acumulado
basePop['% acum'] = basePop['%'].cumsum()

basePop
```

```python
#Formatar percentual
basePop['%'] = basePop['%'].map("{:0.1%}".format)
basePop['% acum'] = basePop['% acum'].map("{:0.1%}".format)

basePop
```
